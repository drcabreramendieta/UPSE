```python
import pandas as pd
import numpy as np
```

```python
df_TR = pd.read_excel('../datos.xlsx', sheet_name='TR', header=None)
df_TR
```

```python
df_TF = pd.read_excel('../datos.xlsx', sheet_name='TF', header=None)
df_TF
```

```python
a_TR = df_TR.to_numpy()
a_TF = df_TF.to_numpy()
a_TR, a_TF
```

```python
x_TR = a_TR[:,1:-1]
x_TF = a_TF[:,1:-1]
y = a_TR[:,-1]
```

```python
x_TR
```

```python
x_TF
```

```python
X = np.array([x_TR,x_TF]).transpose(1,0,2)
```

```python
X.shape
```

```python
y.shape
```

```python
import torch
```

```python
X_T=torch.tensor(X).float()
```

```python
X_T
```

```python
X_T.shape, X_T.dtype
```

```python
X_T
```

```python
ds=torch.utils.data.TensorDataset(X_T, torch.tensor(y).long())
ds
```

```python
from torch.utils.data.dataset import random_split
train_val_ds, test_ds = random_split(ds, [0.8,0.2])
```
